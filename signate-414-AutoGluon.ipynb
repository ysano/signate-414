```
import pandas as pd
import numpy as np
from autogluon.tabular import TabularDataset, TabularPredictor
```

```
directory = '/content/drive/MyDrive/Colab Notebooks/signate-414/'
label = 'Outcome'
eval_metric = 'accuracy'
save_path = directory + 'AutoGluonModels/'
```

```
train_data = TabularDataset(directory + 'train.csv')
subsample_size = 500  # subsample subset of data for faster demo, try setting this to much larger values
#train_data = train_data.sample(n=subsample_size, random_state=0)
train_data.head()
```

```
print("Summary of class variable: \n", train_data[label].describe())
```

```
predictor = TabularPredictor(label=label, problem_type='binary', eval_metric=eval_metric, path=save_path, verbosity=3, learner_kwargs={'ignored_columns': 'index'}).fit(
    train_data, presets='best_quality',
    time_limit=7200,
    hyperparameter_tune_kwargs='auto',
)

results = predictor.fit_summary()
```

```
test_data = TabularDataset(directory + 'test.csv')
test_data.head()
```

```
predictor.positive_class
```

```
predictor = TabularPredictor.load(save_path)  # unnecessary, just demonstrates how to load previously-trained predictor from file
y_pred = predictor.predict(test_data)
```

```
sample_submit = pd.read_csv(directory+'sample_submit.csv', header=None)
submission = pd.concat([sample_submit.iloc[:,0], y_pred], axis=1)
submission.to_csv(directory+'submit_autogluon-hpo-7200.csv', index=None, header=None)
```
